# Набор данных о сердечно-сосудистых заболеваниях

## Данные о датасете


https://www.kaggle.com/datasets/thedevastator/exploring-risk-factors-for-cardiovascular-diseas

Предметная область - медицина.
Характер данных - реальные.

Атрибуты, представленные в датасете:
1) index - индекс пациента (int)
2) id - индекс пациента (int)
1) age - возраст в днях (int) 
2) height  - рост в см (int)
3) weight - вес  в кг (float) 
4) gender - пол (int, 1 - женщина, 2 - мужчина)
5) ap_hi -  систолическое кровяное давление (int) 
6) ap_lo - диастолическое кровяное давление (int)
7) cholesterol - уровень холестерина (int, 1 - нормальный, 2 - выше нормы, 3 - значительно выше нормы)
8) gluc - уровень глюкозы  (int, 1 - нормальный, 2 - выше нормы, 3 - значительно выше нормы)
9) smoke - курение (int, 0 - не курит, 1 - курит)
10) alco - потребление алкоголя (int, 0 - не употребляет, 1 - потребляет)
11) active - физическая активность (int, 0 - отсутствует, 1 - присутствует)
12) cardio - наличие или отсутствие сердечно-сосудистых заболеваний (int, 0 - отсутствие, 1 - наличие)

Задача иследования:
Предсказание наличия или отсутствия сердечно-сосудистых заболеваний с использованием результатов обследования пациента.

### Тепловая карта корреляции Спирмена
